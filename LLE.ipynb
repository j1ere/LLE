```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_swiss_roll
from sklearn.manifold import LocallyLinearEmbedding
```

```python
#Generating a synthetic dataset
#Generating a synthetic dataset resembling a swiss roll using the make_swiss_roll function from scikit-learn

# n_samples = Specifing the number of datapoints to be generated
# n_neighbors = defining the number of neighbors used in the LLE algorithm

n_samples = 1000
n_neighbors = 10

X, _ = make_swiss_roll(n_samples=n_samples)
```

```python
#applying LLE 
#An instance of the LLE algorithm is created using LocallyLinearEmbedding
#the n_neighbors parameter determines the number of neighbors to consider during the embedding process
#the LLE algorithm is then fitted to the original data X using the fit_transform method reducing dataset
#to two dimensions (n_components=2)

lle = LocallyLinearEmbedding(n_neighbors=n_neighbors, n_components=2)
X_reduced = lle.fit_transform(X)
```

```python
#Visualizing the original and reduced data
#after reducing the dimensions with LLE we plot the original high dimensional data and the new lower dimensional data
#this helps us see if important patterns are preserved and easier to understand

plt.figure(figsize=(12,6))
plt.subplot(111)
plt.scatter(X[:, 0], X[:, 1], c=X[:, 2], cmap=plt.cm.Spectral)
plt.title("Original Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

```

```python
plt.subplot(111)
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=X[:, 2], cmap=plt.cm.Spectral)
plt.title("Reduced Data LLE")
plt.xlabel("Component 1")
plt.ylabel("Component 2")
```

```python
plt.figure(figsize=(12,6))
plt.subplot(121)
plt.scatter(X[:, 0], X[:, 1], c=X[:, 2], cmap=plt.cm.Spectral)
plt.title("Original Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

plt.subplot(122)
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=X[:, 2], cmap=plt.cm.Spectral)
plt.title("Reduced Data LLE")
plt.xlabel("Component 1")
plt.ylabel("Component 2")

plt.tight_layout()
plt.show()
```

```python
# the second subplot shows the LLE-reduced data (X_reduced), colored by the third original feature X[:, 2]
#plt.tight_layout() ensures proper spacing between subplots

#LLE preserves local structure
# LLE has a moderate computational cost
#LLE is best used for nonlinear manifolds
#LLE is good for visualization. Projects data into 2D or 3D spaces without 
#losing much structure hence useful for visualizing high dimensional datasets.
#(Unsupervised) - does not require labels so it works well for exploratory data analysis
#in various domains like images, speech or genetics

# Disadvantages of LLE
#Sensitive to parameter
#No Out-of-Sample Mapping
#Sensitive to noise and outliers
#Loss of Global Structure
#Curse of Dimensionality - in very high dimensional spaces, more neighbors are needed to capture
#local structures which increases computational costs.
```
